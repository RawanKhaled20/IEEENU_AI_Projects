```python
from scipy.stats import pearsonr

data = titanic[['fare', 'age']].dropna()

x = data['fare']
y = data['age']

# Create a joint plot using Seaborn
joint = sns.jointplot(x=x, y=y)

# Calculate Pearson correlation coefficient and p-value
person_corr, p = pearsonr(x, y)

# Annotate the joint plot with the correlation coefficient and p-value
text = f"Pearson Correlation: {person_corr:.3f}   p: {p:.2f}"
joint.ax_joint.annotate(text, xy=(0.5, 0.9), xycoords='axes fraction', fontsize=6)

plt.show()  # Display the plot
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # Create subplots for separate plots

# Filter data for males and females
female_data = titanic[titanic['sex'] == 'female']
male_data = titanic[titanic['sex'] == 'male']

# Male histogram
sns.histplot(data=male_data, x='age', bins=10, ax=axes[0], color='darkblue')
axes[0].set_title('sex=male')

# Female histogram
sns.histplot(data=female_data, x='age', bins=10, ax=axes[1], color='darkblue')
axes[1].set_title('sex=female')


plt.show()
```
