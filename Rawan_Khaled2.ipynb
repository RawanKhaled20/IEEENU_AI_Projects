** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Create a figure object
fig = plt.figure()

# Add axes to the figure canvas at different positions
ax1 = fig.add_axes([0, 0, 1, 1])           # Main plot
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # Inserted plot

# Plot (x, y) on the main plot (ax1)
ax1.plot(x, y, color='r')  # Main plot with red line

# Plot (x, z) on the inserted plot (ax2) with specific x and y limits
ax2.plot(x, z, color='b')  # Inserted plot with blue line
ax2.set_xlim(2, 4)  # Limit x-axis from 2 to 4
ax2.set_ylim(6, 10)  # Limit y-axis from 6 to 10

# Show the figure with both plots on different axes
plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Create a figure with subplots
fig, axs = plt.subplots(nrows=1, ncols=2)

# Plotting on the first subplot (axs[0])
axs[0].plot(x, y, color='r')  # First subplot with red line

# Plotting on the second subplot (axs[1])
axs[1].plot(x, z, color='b')  # Second subplot with blue line
axs[1].set_xlim(2, 4)  # Limit x-axis from 2 to 4
axs[1].set_ylim(6, 10)  # Limit y-axis from 6 to 10


plt.show()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Create a figure with subplots
fig, axs = plt.subplots(nrows=1, ncols=2)

# Plot (x, y) on the first subplot (axs[0])
axs[0].plot(x, y, color='green', linestyle='--', linewidth=2)  # First subplot with green dashed line

# Plot (x, z) on the second subplot (axs[1])
axs[1].plot(x, z, color='blue', linestyle=':', linewidth=3)  # Second subplot with blue dotted line
axs[1].set_xlim(2, 4)  # Limit x-axis from 2 to 4
axs[1].set_ylim(6, 10)  # Limit y-axis from 6 to 10


plt.show()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # Figsize set to (width=10, height=4)

# Plot (x, y) on the first subplot (axs[0])
axs[0].plot(x, y, color='green', linestyle='--', linewidth=2)  # First subplot with green dashed line

# Plot (x, z) on the second subplot (axs[1])
axs[1].plot(x, z, color='blue', linestyle=':', linewidth=3)  # Second subplot with blue dotted line
axs[1].set_xlim(2, 4)  # Limit x-axis from 2 to 4
axs[1].set_ylim(6, 10)  # Limit y-axis from 6 to 10


plt.show()
```
